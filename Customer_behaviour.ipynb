Filling the missing value with the median rating of its category

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#format the column names
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
print(df.columns)
```

# Feature Engineering

```python
# create a colum age_group
labels = ['Young_Adult', 'Adult', 'Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4,labels=labels)
```

```python
df[['age','age_group']].head()
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly': 90,
    'Bi-Weekly' : 14,
    'Annually': 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

# Connecting DataBase

```python
import psycopg2
import json
```

```python
# step 1: Connect with Postgres SQL
config_db = 'db_config.json'
try:
    with open(config_db,'r') as f:
        db_config = json.load(f)
    print(f"config is loaded successfuly from file {config_db}")
except Exception as e:
        print(f"Error occured while reading from json file : {config_db}, Error : {e}")
    
```

```python
DB_NAME = db_config.get("db_name")
USER = db_config.get("user")
PASSWORD = db_config.get("password")
HOST = db_config.get("host")
PORT = db_config.get("port")
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine(f"postgresql+psycopg2://{USER}:{PASSWORD}@{HOST}:{int(PORT)}/{DB_NAME}")
```

```python
# load dataset
tabel_name = 'customer'
df.to_sql(tabel_name, engine, if_exists='replace', index=False)
```
